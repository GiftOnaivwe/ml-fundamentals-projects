#### A DECISION TREE is a popular machine learning algorithm used for both classification and regression tasks. It works by splitting the data into subsets based on the value of input features, creating a tree-like model of decisions.

```python
#import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
#load dataset
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
# Define X and Y, independent and dependent variable
X=df.drop(columns=['Outcome'])
Y=df['Outcome']
```

```python
#Split the data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#Scale the features
scaler=StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
clf=DecisionTreeClassifier()
```

```python
#Train model
clf.fit(X_train,Y_train)
```

```python
#Predict model values
y_pred=clf.predict(X_test)
```

```python
accuracy=accuracy_score(Y_test,y_pred)
accuracy
```

```python
#Visualize tree
plt.figure(figsize=(20,10))
tree.plot_tree(clf,feature_names=X.columns)
plt.title('Decision tree for Diabetes Datasets')
plt.show()
```
