```python
#Probability distribution describes how data is distributed; it i a mathematical functions that provides probabilities of occurence of different possible outcome of an experiment
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
uniform_samples = np.random.uniform(0,1,1000)
plt.figure(figsize=(8,6))
plt.hist(uniform_samples,bins=30,color='maroon',alpha=0.9)
plt.title('Uniform Distribution')
plt.show()
```

# Normal Distribution

```python
#Guassain Distrinution
# data points clustered around mean
#Symmetric; bell-shaped curve
#Mean, Median and Mode equal
# Use mean, std, e
```

```python
normal_samples = np.random.normal(0,1,1000)
plt.figure(figsize=(8,6))
plt.hist(normal_samples,bins=30,color='maroon',alpha=0.9)
plt.title('Normal Distribution')
plt.show()
```

# Binomial Distribution

```python
#.5 means the number of probability of success
binomial_samples = np.random.binomial(10,0.5,1000)
plt.figure(figsize=(8,6))
plt.hist(binomial_samples,bins=30,color='maroon',alpha=0.9)
plt.title('Binomial Distribution')
plt.show()
```

# Poisson Distribution

```python
#Poisson distribution counts the number of events. eg number of cars passing through tolls, number of triggers events. 
#In summary average number of events
```

```python
poisson_samples = np.random.poisson(5,1000)
plt.figure(figsize=(8,6))
plt.hist(poisson_samples,bins=30,color='black',alpha=0.9)
plt.title('Poisson Distribution')
plt.show()
```
